# Examples of S curves

Example of a [logistic function](https://en.wikipedia.org/wiki/Logistic_function)

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x, L, k, x0):
    return L / (1 + math.exp(-k * (x - x0)))

x = np.linspace(-30, 30, 100)

v_f = np.vectorize(f)

y = v_f(x, 1, 0.3, 0)

plt.plot(x, y, color='green')

plt.show()
```

Same logistic function but with output shifted in `t` and with different `k` values

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x, L, k, x0):
    return L / (1 + math.exp(-k * (x - x0)))

max_x = 60

x = np.linspace(0, max_x, 100)

v_f = np.vectorize(f)

y0 = v_f(x - max_x / 2, 1, 0.1, 0)
y1 = v_f(x - max_x / 2, 1, 0.2, 0)
y2 = v_f(x - max_x / 2, 1, 0.3, 0)
y3 = v_f(x - max_x / 2, 1, 0.7, 0)
y4 = v_f(x - max_x / 2, 1, 1.0, 0)

plt.plot(x, y0, color='yellow')
plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='green')
plt.plot(x, y3, color='red')
plt.plot(x, y4, color='orange')

plt.show()
```

```python
import math
from matplotlib import pyplot as plt
import numpy as np


def y(x):
    a = 1.0
    b = 1.0
    c = 1.0
    y_squared = (c - a * x * x * x) / b
    if y_squared > 0:
        return math.sqrt(y_squared)
    else:
        return -math.sqrt(abs(y_squared))


v_y = np.vectorize(y)

x = np.arange(0.0, 1.0, 1.0 / 30)

plt.plot(x, v_y(x), 'g^')
plt.show()
```
